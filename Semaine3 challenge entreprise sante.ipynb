```python
#====================================================================================================
# CHALLENGE ENTREPRISE - DEMANDE DU MINISTRE : RAPPORT D'ETAT SANITAIRE DEPARTEMENTAL
#====================================================================================================

# DECLARATION DES VARIABLES 
# Hopital CHU de Brazzaville

hopital1 = {
    'lits_totaux': 320, 'lits_occupes' : 298, 'nb_medecins': 47, 'nb_ruptures': 2, 'nom_rupture' : ('SRO, Vaccin') , 
    'nb_alertes': 2, 'nom_alerte': ('Artemether,Amoxi') 
}
# Hopital de Pointe-Noire

hopital2 = {
    'lits_totaux': 180, 'lits_occupes' : 143,'nb_medecins': 22, 'nb_ruptures': 0, 
    'nb_alertes': 1, 'nom_alerte': ('Amoxicilline')
}
# Hopital de Dolisie

hopital3 = {
    'lits_totaux': 95, 'lits_occupes' : 91,'nb_medecins': 8, 'nb_ruptures': 1, 'nom_rupture' :('SRO'), 
    'nb_alertes': 2, 'nom_alerte': ('Artemether, Vaccin')
}
#Hopital de Owando
hopital4 = {
    'lits_totaux':45, 'lits_occupes' : 32,'nb_medecins':  3, 'nb_ruptures': 3, 'nom_ruptue': ('SRO, Vaccin, Arthemer'),
    'nb_alertes': 0 
}
#CMS de Impfondo
hopital5 = { 
    'lits_totaux': 20 , 'lits_occupes' : 19,'nb_medecins': 1, 'nb_ruptures': 2, 'nom_ruptue':('SRO, Amoxi'),
    'nb_alertes': 1, 'nom_alerte':('Vaccin')
}
# Calcule des taux

hopitaux = [hopital1, hopital2, hopital3, hopital4, hopital5]
for hopital in hopitaux:
    hopital['taux_occupation'] = round((hopital["lits_occupes"]/hopital["lits_totaux"])*100,1)

def niveau_alerte_global(hopital):
    if hopital['nb_ruptures'] >= 2 or hopital['taux_occupation'] > 95:
        return "CRITIQUE"
    elif hopital["nb_ruptures"] >= 1 or hopital['taux_occupation'] > 85 or (hopital['nb_alertes'] >= 2 and hopital['nb_medecins'] < 5):
        return "PREOCCUPANT"
    else:
        return "SATISFAISANT"
    
print("="*70)
print("TABLEAU DE BORD SANITAIRE - MINISTRE DE LA SANTE ")
print("Date : 16 janvier 2026 | Pour le Conseil des Ministres")
print("="*70)
print("HOPITAL                   OCCUPATION     ALERTES      NIVEAU GLOBAL ")
print(f'CHU Brazzaville           {hopital1['taux_occupation']}% [ALT]    2R + 2A      [{niveau_alerte_global(hopital1)}]')
print(f'Hopital Pointe-Noire      {hopital2['taux_occupation']}% [OK]     0R + 1A      [{niveau_alerte_global(hopital2)}]')
print(f'Hopital dolisie           {hopital3['taux_occupation']}% [CRI]    1R + 2A      [{niveau_alerte_global(hopital3)}]')
print(f'Hopital Owando            {hopital4['taux_occupation']}% [OK]     3R + 0A      [{niveau_alerte_global(hopital4)}]')
print(f'CMS Impondo               {hopital5['taux_occupation']}% [ALT]    2R + 1A      [{niveau_alerte_global(hopital5)}]')
print("-"*70)
print("4 hopitaux sur 5 en situation CRITIQUE")
print("8 ruptures de stocks identifiees a l'echelle nationale")
print("RECOMMANDATION PRIORITAIRE : Mobiliser la reserve nationale PNA")
print("="*70)
```

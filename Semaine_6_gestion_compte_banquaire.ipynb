```python
# Gestion d'un compte Bancaire
class CompteBancaire:

    def __init__(self, numero, titulaire, pin, solde=0):
        self.numero = numero
        self.titulaire = titulaire
        self.pin = pin
        self.solde = solde
        self.historique = []

    def deposer(self, montant):

        if montant > 0:
            self.solde += montant
            self.historique.append(f"Dépôt : +{montant} FCFA")
            print("Dépôt effectué.")
        else:
            print("Montant invalide.")

    def retirer(self, montant):

        if montant <= 0:
            print("Montant invalide.")

        elif montant > self.solde:
            print("Solde insuffisant.")

        else:
            self.solde -= montant
            self.historique.append(f"Retrait : -{montant} FCFA")
            print("Retrait effectué.")

    def afficher_solde(self):
        print(f"Solde : {self.solde:.2f} FCFA")

    def informations(self):

        print("\n========== INFORMATIONS ==========")
        print("Titulaire :", self.titulaire)
        print("Compte :", self.numero)
        print("Solde :", self.solde, "FCFA")

    def afficher_historique(self):

        print("\n===== HISTORIQUE =====")

        if len(self.historique) == 0:
            print("Aucune opération.")

        else:

            for operation in self.historique:
                print(operation)


banque = {}


def creer_compte():

    numero = input("Numéro du compte : ")

    if numero in banque:
        print("Compte déjà existant.")
        return

    titulaire = input("Nom du titulaire : ")
    pin = input("Créer un code PIN : ")

    try:
        solde = float(input("Dépôt initial : "))
    except ValueError:
        print("Montant invalide.")
        return

    banque[numero] = CompteBancaire(numero, titulaire, pin, solde)

    print("Compte créé avec succès.")


def connexion():

    numero = input("Numéro du compte : ")
    pin = input("Code PIN : ")

    if numero not in banque:
        print("Compte introuvable.")
        return None

    compte = banque[numero]

    if compte.pin != pin:
        print("PIN incorrect.")
        return None

    print("Connexion réussie.")
    return compte


def virement(compte):

    destinataire = input("Compte destinataire : ")

    if destinataire not in banque:
        print("Compte inexistant.")
        return

    try:
        montant = float(input("Montant : "))
    except ValueError:
        print("Montant invalide.")
        return

    if montant <= 0:
        print("Montant invalide.")
        return

    if montant > compte.solde:
        print("Solde insuffisant.")
        return

    compte.solde -= montant
    banque[destinataire].solde += montant

    compte.historique.append(
        f"Virement envoyé : {montant} FCFA vers {destinataire}"
    )

    banque[destinataire].historique.append(
        f"Virement reçu : {montant} FCFA de {compte.numero}"
    )

    print("Virement effectué.")


while True:

    print("\n===============================")
    print("      BANQUE PYTHON")
    print("===============================")
    print("1. Créer un compte")
    print("2. Se connecter")
    print("3. Quitter")

    choix = input("Choix : ")

    if choix == "1":
        creer_compte()

    elif choix == "2":

        compte = connexion()

        if compte is None:
            continue

        while True:

            print("\n===== MENU CLIENT =====")
            print("1. Déposer")
            print("2. Retirer")
            print("3. Consulter le solde")
            print("4. Informations")
            print("5. Historique")
            print("6. Virement")
            print("7. Déconnexion")

            choix2 = input("Choix : ")

            if choix2 == "1":

                try:
                    montant = float(input("Montant : "))
                    compte.deposer(montant)
                except ValueError:
                    print("Montant invalide.")

            elif choix2 == "2":

                try:
                    montant = float(input("Montant : "))
                    compte.retirer(montant)
                except ValueError:
                    print("Montant invalide.")

            elif choix2 == "3":
                compte.afficher_solde()

            elif choix2 == "4":
                compte.informations()

            elif choix2 == "5":
                compte.afficher_historique()

            elif choix2 == "6":
                virement(compte)

            elif choix2 == "7":
                print("Déconnexion...")
                break

            else:
                print("Choix invalide.")

    elif choix == "3":
        print("Merci d'avoir utilisé Banque Python.")
        break

    else:
        print("Choix invalide.")
```

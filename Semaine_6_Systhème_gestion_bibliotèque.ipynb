```python
# SYSTEME DE GESTION DE BIBLIOTHEQUE

bibliotheque = {
    "Python": True,
    "Machine Learning": True,
    "Économétrie": True,
    "Statistiques": True
}

while True:

    print("\n====== BIBLIOTHÈQUE ======")
    print("1. Afficher les livres")
    print("2. Emprunter un livre")
    print("3. Retourner un livre")
    print("4. Ajouter un livre")
    print("5. Quitter")

    choix = input("Votre choix : ")

    if choix == "1":

        print("\nListe des livres :")

        for livre, disponible in bibliotheque.items():

            if disponible:
                etat = "Disponible"
            else:
                etat = "Emprunté"

            print(f"{livre} --> {etat}")

    elif choix == "2":

        livre = input("Nom du livre : ")

        if livre in bibliotheque:

            if bibliotheque[livre]:
                bibliotheque[livre] = False
                print("Livre emprunté.")
            else:
                print("Livre déjà emprunté.")

        else:
            print("Livre introuvable.")

    elif choix == "3":

        livre = input("Nom du livre : ")

        if livre in bibliotheque:
            bibliotheque[livre] = True
            print("Livre retourné.")
        else:
            print("Livre inexistant.")

    elif choix == "4":

        livre = input("Nom du nouveau livre : ")

        if livre in bibliotheque:
            print("Le livre existe déjà.")
        else:
            bibliotheque[livre] = True
            print("Livre ajouté.")

    elif choix == "5":
        print("Merci d'avoir utilisé la bibliothèque.")
        break

    else:
        print("Choix invalide.")
```
